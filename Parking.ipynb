```python
import pandas as pd
import numpy as np
```

```python
#read the file
df = pd.read_csv('311_Service_Requests_2020.csv')
```

```python
#deal with zip code
def fix_zip(input_zip):
    try:
        input_zip = int(float(input_zip))
    except:
        try:
            input_zip = int(input_zip.split('-')[0])
        except:
            return np.NaN
    if input_zip < 10000 or input_zip > 12000:
        return np.NaN
    return str(input_zip)
        
```

```python
df['Incident Zip'] = df['Incident Zip'].apply(fix_zip)
```

```python
df['Incident Zip'].unique()
```

```python
df = df[df['Incident Zip'].notnull()]
```

```python
new_df = df[df['Incident Zip']=='10025']
```

```python
new_df.iloc[0:10]
```

```python
new_df['Complaint Type'].unique()
```

```python
#number of illegal parking
illegal_zip = new_df[new_df['Complaint Type'] == 'Illegal Parking'].count()['Incident Zip']
total = new_df.count()['Complaint Type']
fraction = illegal_zip / total
print(fraction)
```

```python
#total number of illegal parking
illegal = df[df['Complaint Type'] == 'Illegal Parking'].count()['Incident Zip']
total_number = df.count()['Incident Zip']
fraction_total = illegal / total_number
print(fraction_total)
```

```python
#boolean
if fraction > fraction_total:
    higher_parking_portation = True
else:
    higher_parking_portation = False
print(higher_parking_portation)
```
